```python
import json
```

```python
def load_poetry(fn: str):
    with open(fn, 'rt', encoding='utf-8') as f:
        poetry = json.loads(f.read())
    return poetry
```

```python
classics = load_poetry('classics.json')
print("Authors: ", len(classics))
print("Poems: ", sum(len(poems) for author, poems in classics.items()))
```

```python
modern = load_poetry('modern.json')
print("Authors: ", len(modern))
print("Poems: ", sum(len(poems) for author, poems in modern.items()))
```

```python
def dic2list(dic):
    return [(k, v) for k, v in dic.items()]
```

```python
classics_list = [(author, dic2list(poems)) for author, poems in classics.items()]
```

```python
print(classics_list[5][1][3][1])
```

```python
modern_list = [(author, dic2list(poems)) for author, poems in modern.items()]
```

```python
print(modern_list[5][1][2][1])
```

```python
common_list = classics_list+modern_list
```

```python
with open('all.txt', 'wt', encoding='utf-8') as f:
    for author, poems in common_list:
        for title, poem in poems:
            lines = poem.splitlines()
            lines = lines[1:]
            f.write('\n'.join(lines))
            f.write('\n')
        
```
